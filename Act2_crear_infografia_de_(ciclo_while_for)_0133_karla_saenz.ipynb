```python
# IF SIMPLE
# Karla Saenz 0133
numero = int(input("Ingresa un número: "))

if numero % 2 == 0:
    print("El número es par")
```

```python
# IF ELSE

numero = int(input("Ingresa un número: "))

if numero % 2 != 0:
    print("El número es impar")
else:
    print("El número es par")
    print("elaborado por karla saenz0133")
```

```python
# PUEDE VOTAR

edad = int(input("Ingresa tu edad: "))

if edad >= 18:
    print("Puedes votar")
else:
    print("No puedes votar")
```

```python
# CALIFICACIONES

nota = int(input("Ingresa tu calificación: "))

if nota >= 90:
    print("Excelente")

elif nota >= 70:
    print("Bueno")

else:
    print("Necesita mejorar")
```
